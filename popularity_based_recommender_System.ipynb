```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('/content/drive/MyDrive/DataSets/BX-Books.csv',sep = ';',encoding='latin-1',error_bad_lines=False ,names = ['ISBN', 'title', 'author', 'year', 'publisher','Image-1','Image-2','Image-3'])
ratings = pd.read_csv('/content/drive/MyDrive/DataSets/BX-Book-Ratings.csv',sep = ';',encoding='latin-1',error_bad_lines=False)
users = pd.read_csv('/content/drive/MyDrive/DataSets/BX-Users.csv',sep = ';',encoding='latin-1',error_bad_lines=False)
```

```python
rating_with_name = ratings.merge(books,on = 'ISBN')
rating_with_name
```

```python
num_rating_df = rating_with_name.groupby('title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'number_of_ratings'},inplace = True)
```

```python
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg_of_ratings'},inplace = True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on = 'title')
```

```python
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['number_of_ratings']>=250].sort_values('avg_of_ratings',ascending = False).head(50)
```
