### Custom Constraints
Consider only important users(rated atleast 200 books) and important books(rated by atleast 50 users)

```python
## True for important users else False
```

```python
x = ratings['user_id'].value_counts()>200
x.value_counts()
```

```python
## index of only True value
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

### Important Books Filtering

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns ={'rating' : 'number_of_ratings'},inplace = True)
```

```python
final_rating = ratings_with_books.merge(number_rating , on = 'title')
```

```python
final_rating.columns
```

```python
final_rating
```

```python
final_rating = final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating.columns
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace = True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id' ,index = 'title' , values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace = True)
```

```python
book_pivot
```

```python
## To handle Sparse Matrix (Scientific Python)
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model =  NearestNeighbors(algorithm ='brute')
```

```python
model.fit(book_sparse)
```

```python
def recommendation(book_id,n):
  distance,suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=n)
  return suggestions
```

```python
### Deriver Code
book_id = 237
recommendation_number = 6
suggestions = recommendation(book_id, recommendation_number)
pd.DataFrame(book_pivot.index[suggestions])
```
